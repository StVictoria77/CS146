```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt

# Field goal rate data. Each tuple contains (field goals made, field goals attempted).
player_field_goal_data = {
    # http://www.espn.com/nba/player/gamelog/_/id/2594922/year/2018/otto-porter-jr
    'Otto Porter Jr': [9, 8, 7, 10],
    # http://www.espn.com/nba/player/gamelog/_/id/3024/year/2018/jj-redick
    'JJ Redick': [13, 11, 15, 17],
    # http://www.espn.com/nba/player/gamelog/_/id/3213/year/2018/al-horford
    'Al Horford': [12, 8, 9, 13],
}
```

```python
for player in player_field_goal_data.keys():
    
    # Prior: Beta distribution
    alpha_prior = 1
    beta_prior = 1
    prior_field_goal_rate = sts.beta.rvs(alpha_prior, beta_prior, size = 1)
    
    goals = []

    for n in player_field_goal_data[player]:
        k = sts.binom.rvs(n, prior_field_goal_rate, size = 1)
        goals.append(k)

    print('Posterior over FG% for ' + player, [int(k) for k in goals], 'given the prior field goal rate of', prior_field_goal_rate)
```
